<h2>线图</h2>

<h3>折线曲线图</h3>

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.faker import Faker
c = Line()
c.add_xaxis(Faker.choose())
c.add_yaxis("商家1", Faker.values(), is_smooth=True, color="red")
c.add_yaxis("商家2", Faker.values(), is_smooth=False, color="blue")
c.set_global_opts(title_opts=opts.TitleOpts(
    title="主标题"
))
c.render_notebook()
```

<h3>设置标记点</h3>

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.faker import Faker
x, y = Faker.choose(), Faker.values()
c = Line()
c.add_xaxis(x)
c.add_yaxis(
    "商家1",
    y,
    markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(
        name="自定义标记点",
        coord=(x[2], y[2]),
        value=y[2]
    )])
)
c.set_global_opts(title_opts=opts.TitleOpts(
    title="主标题"
))
c.render_notebook()
```

<h3>设置标记线</h3>

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.faker import Faker
c = Line()
c.add_xaxis(Faker.choose())
c.add_yaxis("商家1", Faker.values(),
    markline_opts=opts.MarkLineOpts(data=[
        opts.MarkLineItem(type_="average", name="平均值")
    ])
)
c.add_yaxis("商家2", Faker.values(),
    markline_opts=opts.MarkLineOpts(data=[
        opts.MarkLineItem(type_="max", name="最大值")
    ])
)
c.add_yaxis("商家2", Faker.values(),
    markline_opts=opts.MarkLineOpts(data=[
        opts.MarkLineItem(type_="median", name="中位数")
    ])
)
c.set_global_opts(title_opts=opts.TitleOpts(
    title="主标题"
))
c.render_notebook()
```

<h3>设置线形</h3>

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.faker import Faker
c = Line()
c.add_xaxis(xaxis_data=Faker.choose())
c.add_yaxis("商家A", Faker.values(), symbol="triangle", symbol_size=20,
        linestyle_opts=opts.LineStyleOpts(color="green", width=4, type_="dashed"),
        itemstyle_opts=opts.ItemStyleOpts(border_width=3, border_color="yellow", color="blue"))
c.set_global_opts(title_opts=opts.TitleOpts(title="主标题"))
c.render_notebook()
```

<h2>饼图</h2>

```python
import pyecharts.options as opts
from pyecharts.charts import Pie
from pyecharts.faker import Faker
c = Pie()
c.add("", 
      [list(z) for z in zip(Faker.choose(), Faker.values())],  # 填入数据
      radius=["40%", "60%"]  # 设置内径外径
)
c.set_global_opts(
    title_opts=opts.TitleOpts(title="标题"),  # 设置标题
    legend_opts=opts.LegendOpts(orient="vertical", pos_top="15%", pos_left="2%")  # 设置图例位置
)
c.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))  # 设置tooltip
c.set_colors(["blue", "green", "yellow", "red", "pink", "orange", "purple"])  # 设置颜色
c.render_notebook()
```

<h2>散点图</h2>

```python
import pyecharts.options as opts
from pyecharts.charts import Scatter

data = [[10.0, 8.04],[8.0, 6.95],[13.0, 7.58],[9.0, 8.81],[11.0, 8.33],[14.0, 9.96], [6.0, 7.24],[4.0, 4.26],[12.0, 10.84],[7.0, 4.82],[5.0, 5.68]]
nd = np.array(data)
x_data = nd[:, 0]
y_data = nd[:,1]
s = Scatter()
s.add_xaxis(xaxis_data=x_data)  # 填入x数据
s.add_yaxis(
    series_name="",
    y_axis=y_data,  # 填入y数据
    symbol_size=5,  # 点大小
    label_opts=opts.LabelOpts(is_show=True)  # 是否显示点上方数字
)
s.set_global_opts(
        xaxis_opts=opts.AxisOpts(type_="value", splitline_opts=opts.SplitLineOpts(is_show=True)),  # 是否显示横向grid
        yaxis_opts=opts.AxisOpts( type_="value",
            axistick_opts=opts.AxisTickOpts(is_show=True),  
            splitline_opts=opts.SplitLineOpts(is_show=True)  # 是否显示纵向grid
        ),   
        tooltip_opts=opts.TooltipOpts(is_show=True),  # 是否开启Tooltip
    )
s.render_notebook()
```
