<h2>散点图</h2>

<h3>plot方法</h3>

<h4>圆点标记</h4>

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(1)
plt.rcParams['font.sans-serif']=['SimHei']
x = np.array([6, 8, 10, 14, 18])
y1 = np.array([7, 9, 13, 17.5, 18])
y2 = y1[::-1]
y3 = y1 + np.max(y1)
y4 = y2 + np.max(y2)
plt.plot(x, y1, "b.", label="点标记")
plt.plot(x, y2, "ro", label="实心圆")
plt.plot(x, y3, "k,", label="极小点")
plt.plot(x, y4, "go", markerfacecolor="none", label="空心圆")
plt.xlabel("横轴")
plt.ylabel("纵轴")
plt.title("标题")
plt.grid(True)
plt.legend()
plt.show()
```

<h3>scatter方法</h3>

<h4>画一类点</h4>

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
area = (30 * np.random.rand(50)) ** 2
plt.figure(1)
plt.rcParams['font.sans-serif']=['SimHei']
plt.scatter(
    x,  # 横坐标
    y,  # 纵坐标
    s=area,  # 面积
    c=colors, # 颜色
    alpha=0.2 # 透明度
)
plt.show()
```

<h4>每个类一种颜色</h4>

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(1)
plt.rcParams['font.sans-serif']=['SimHei']
colors = ["tab:red", "tab:orange", "tab:blue"]
for color in colors:
    x, y = np.random.rand(2, 50)
    area = (30 * np.random.rand(50)) ** 2
    plt.scatter(
        x, 
        y,
        c=color,
        s=area,
        alpha=0.2,
        label=color,
        edgecolors=colors[(colors.index(color) + 1) % len(colors)]
    )
plt.legend()
plt.grid(True)
plt.show()
```

<h2>柱状图</h2>

<h3>堆叠柱形图</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(1)
plt.rcParams['font.sans-serif']=['SimHei']
x = np.array(["HUAWEI", "Tencent", "ByteDance", "Baidu", "Alibaba"])
y1 = np.array([7, 9, 13, 17.5, 18])
y2 = y1[::-1]
y3 = y1 + np.max(y1)
y4 = y2 + np.max(y2)
plt.bar(x, y1, yerr=np.std(y1), label="money")
plt.bar(x, y3, bottom=y1, yerr=np.std(y3), label="people")
plt.legend()
plt.show()
```

<h3>并列柱形图</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(1)
plt.rcParams['font.sans-serif']=['SimHei']
width = 0.3
y1 = np.array([7, 9, 13, 17.5, 18])
y2 = y1[::-1]
y3 = y1 + np.max(y1)
y4 = y2 + np.max(y2)
x = np.arange(len(y1))
plt.bar(x - width/2, y1, yerr=np.std(y1), width=width, label="money")
plt.bar(x + width/2, y3, yerr=np.std(y3), width=width, label="people")
plt.xticks(x, ["HUAWEI", "Tencent", "ByteDance", "Baidu", "Alibaba"])
plt.legend()
plt.show()
```

<h2>画子图</h2>

<h3>画一张子图</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 500)
y = np.sin(x)
plt.figure(1)
plt.rcParams['font.sans-serif']=['SimHei']
fig, ax = plt.subplots()
ax.plot(x, y, "y-.", label="y=sin(x)")
ax.plot(x, y - 0.2, "r--", label="y=sin(x)-0.2")
ax.set_title("正弦")
ax.set_xlabel("x轴")
ax.set_ylabel("y轴")
ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)
ax.set_xticks(range(11))
ax.set_yticks((-1.5, -1, -0.5, 0, 0.5, 1, 1.5))
ax.legend()
plt.show()
```

<h3>画多张子图</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
data1, data2, data3, data4 = np.random.randn(4, 100)
plt.figure(1)
plt.rcParams['font.sans-serif']=['SimHei']
fig, ((ax00, ax01), (ax10, ax11)) = plt.subplots(2, 2)

ax00.plot(data1, data2, "x")
ax00.set_title("subplot00")
ax00.set_xlabel("x轴")
ax00.set_ylabel("y轴")

ax01.plot(data3, data4, "o")
ax01.set_title("subplot01")
ax01.set_xlabel("x轴")
ax01.set_ylabel("y轴")

ax10.plot(data1, data3, "*")
ax10.set_title("subplot10")
ax10.set_xlabel("x轴")
ax10.set_ylabel("y轴")

ax11.plot(data2, data4, "+")
ax11.set_title("subplot11")
ax11.set_xlabel("x轴")
ax11.set_ylabel("y轴")

plt.tight_layout()
plt.suptitle("标题")
plt.show()
```

<h3>共享x轴</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 7, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(1)
plt.rcParams['font.sans-serif']=['SimHei']
fig, ax1 = plt.subplots()
ax1.plot(x, y1, "--", color="tab:red")
ax1.tick_params(axis="y", labelcolor="tab:red")
ax1.set_ylabel("y1=sin(x)", color="tab:red")

ax2 = ax1.twinx()
ax2.plot(x, y2 + 1, ":", color="tab:blue")
ax2.tick_params(axis="y", labelcolor="tab:blue")
ax2.set_ylabel("y2=cos(x)", color="tab:blue")
plt.show()
```
