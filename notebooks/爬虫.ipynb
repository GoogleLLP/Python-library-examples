<h1>BeautifulSoup库</h1>

<h2>导入库</h2>

```python
import requests
from bs4 import BeautifulSoup
```

<h2>确定爬取网址</h2>

```python
url = "http://www.sohu.com"
```

<h2>获取响应</h2>

```python
response = requests.get(url)
source = response.text
print(source)
```

<h2>爬取内容</h2>

<h3>构建BeautifulSoup对象</h3>

```python
soup = BeautifulSoup(source,"html.parser")       #调用BeautifulSoup类传入源代码source
```

<h3>选择标签</h3>

```python
target_zone = soup.select("div")     #选择器确定为td.al，然后调用soup对象的select方法获得目标内容HTML区
```

<h3>解析标签</h3>

```python
target_texts = [item.text.strip() for item in target_zone if len(item.text.strip()) > 0]  #遍历所有的目标HTML区，将text文本提取后使用列表推导式存入列表
print(target_texts)               #输出实时列表内容
```
