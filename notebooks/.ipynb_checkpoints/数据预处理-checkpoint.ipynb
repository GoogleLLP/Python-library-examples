<h2>iris花案例</h2>

<h3>读取数据</h3>

```python
import numpy as np
data = np.loadtxt("../data/iris.csv", delimiter=",", skiprows=0)
X = data[:, :-1]
y = data[:, -1]
data.shape
```

<h3>平均值、标准差、方差、最大值、最大值位置</h3>

<h4>平均值</h4>

```python
print(np.mean(X[:, 0]))
```

<h4>标准差</h4>

```python
print(np.std(X[:, 0]))
```

<h4>方差</h4>

```python
print(np.var(X[:, 0]))
```

<h4>最大值及其出现的位置</h4>

```python
print(np.max(X[:, 0]))
print(np.argmax(X[:, 0]))
```

<h4>排序</h4>

```python
print(np.sort(X[:, 0]))
```

<h3>查看标签</h3>

```python
print(np.unique(y))
print(y.shape)
print(y[y == 0].shape)
print(y[y == 1].shape)
print(y[y == 2].shape)
```

<h3>数据预处理</h3>

<h4>数据归一化</h4>

```python
X_norm = (X - np.min(X, axis=0)) / (np.max(X, axis=0) - np.min(X, axis=0))
X_norm
```

<h4>数据标准化</h4>

```python
X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
X_norm
```

<h2>diabetes案例</h2>

```python
import pandas as pd
data = pd.read_csv("../data/diabetes.csv")
X = data.iloc[:, :-1]
y = data["Outcome"]
```

<h3>查看数据量</h3>

```python
print(X.shape)
print(y[y == 0].shape)
print(y[y == 1].shape)
```

<h3>查看第0列特征</h3>

<h4>平均值</h4>

```python
X.iloc[:, 0].mean()
```

<h4>查看独一无二值</h4>

```python
np.sort(X.iloc[:, 0].unique())
```

<h4>分组与聚合</h4>

```python
X.groupby(["Pregnancies"]).count()
```

```python
X.iloc[:, 0].value_counts(ascending=False)
```
